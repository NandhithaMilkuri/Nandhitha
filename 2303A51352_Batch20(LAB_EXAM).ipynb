```python
import pandas as pd
data = pd.read_csv('NPHA-doctor-visits.csv')
data.head()
```

```python
top_reasons = data[['Phyiscal Health', 'Mental Health', 'Dental Health']].mean().sort_values(ascending=False)
print("Top 5 Reasons for Healthy Aging:")
print(top_reasons.head(5))
```

```python
community_aging_issues = data.groupby('Race')[['Phyiscal Health', 'Mental Health']].mean().mean(axis=1)
community_high_aging_issues = community_aging_issues.idxmin()
print("\nCommunity with the Highest Aging Issues:", community_high_aging_issues)
```

```python
gender_aging_issues = data.groupby('Gender')[['Phyiscal Health', 'Mental Health']].mean().mean(axis=1)
most_aging_gender = gender_aging_issues.idxmin()
least_aging_gender = gender_aging_issues.idxmax()
print("\nGender with Most Aging Issues:", most_aging_gender)
print("Gender with Least Aging Issues:", least_aging_gender)
```

```python
correlations = data[['Age', 'Employment', 'Mental Health']].corrwith(data['Phyiscal Health'])
print("\nCorrelations with Healthy Aging:")
print(correlations)

```

```python
most_doctor_visits = data.groupby('Age')['Number of Doctors Visited'].sum().idxmax()
print("\nCategory with Most Doctor Visits:", most_doctor_visits)
```
